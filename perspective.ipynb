### Toxicity detection for religion-based identity

```python
# Perspective API Key and Authentication 
API_KEY = 'AIzaSyB2F2u_2RuzpomhwP2ZZucF37q-wW4OWVw'

client = discovery.build(
  "commentanalyzer",
  "v1alpha1",
  developerKey=API_KEY,
  discoveryServiceUrl="https://commentanalyzer.googleapis.com/$discovery/rest?version=v1alpha1",
  static_discovery=False,
)

#comment = input("Enter a sentence: ")
comment=dataframe['Write a sentence that has a word "muslim" in it:']
data= [comment]
terms = read_csv("hatebase.csv")

#selection of a few religion. the code change "muslim" to any of these religious terms and see the toxicity score 
religion=["muslim", "christian", "jew", "athiest", "budhist", "hindu", "bahai", "zoroastrian", "sikh"]
religion_length=len(religion)

new = []
api_output=[]
user_input=[]

for i in data[0]:
    for j in religion:
        new.append(re.sub("muslim",j, i))
        analyze_request = {
            'comment': { 'text': new[-1] },
            'requestedAttributes': {'TOXICITY': {}}
        }
        response = client.comments().analyze(body=analyze_request).execute()
        user_input.append((json.dumps(analyze_request['comment']['text'], indent=0)))
        api_output.append((json.dumps(response["attributeScores"]["TOXICITY"]["summaryScore"]["value"], indent=2)))
        df = pd.DataFrame({'comment':user_input, 'score':api_output})   
new = list(set(new))
api_output= list(set(api_output))
user_input= list(set(user_input))
display(df)
#create a csv file with all the sentences and their scores -- in addition to comparing scores, this can be used as a datset for other hate detection ML models.
df.to_csv('output_religion.csv',index=False)
```
